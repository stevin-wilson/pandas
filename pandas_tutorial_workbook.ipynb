## Filtering data

```python
selected_pokemon = pokemon.loc[((pokemon['Type 1'] == 'Grass') & (pokemon['Type 2'] == 'Poison')) | (pokemon['Attack'] > 80)] # for or , use |
```

```python
selected_pokemon.head()
```

```python
selected_pokemon.reset_index(drop = True, inplace = True)
```

```python
selected_pokemon.head()
```

```python
# select rows with Mega in name
pokemon.loc[pokemon['Name'].str.contains('Mega')]
```

```python
#Remove rows with mega in Name. Can also use Regex
pokemon.loc[-pokemon['Name'].str.contains('Mega')]
```

```python
import re
```

```python
#Use reguar expression
pokemon.loc[pokemon['Name'].str.contains('^Q|^Z*a', regex = True)] # use flag = re.I to ignore case
```

```python
# Change fire in type 1 rows to Flamer
pokemon.loc[pokemon['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
```

```python
pokemon.head(20)
```

```python
# conditional changes

pokemon.loc[pokemon['Total'] > 500, ['Generation', 'Legendary']] = 'TEST VALUE'
```

```python
pokemon.head()
```

```python
pokemon.groupby(['Type 1']).mean().sort_values('Attack')
```

```python
pokemon['count'] = 1
pokemon.groupby(['Type 1']).count()['count']
```

# Large datasets

```python
pd.read_csv('pokemon_data.csv', chunksize = 5)
```

```python
new_pokemon = pd.DataFrame(columns = pokemon.columns[:-2])
```

```python
new_pokemon.head()
```

```python
#read small chunks, group by and concatenate

for df in pd.read_csv('pokemon_data.csv',chunksize = 5):
    results = df.groupby(['Type 1']).count()
    new_pokemon = pd.concat([new_pokemon, results])
```

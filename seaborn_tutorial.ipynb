```python
# Using tutorial from https://seaborn.pydata.org/tutorial.html on pokemon_data.csv

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
from scipy import stats

```

```python
pokemon = pd.read_csv("pokemon_data.csv")
```

```python
pokemon.sample(frac = 0.01)
```

```python
pokemon.loc[pokemon['Type 1'] == 'Rock']
```

```python
sns.relplot(x="Attack", y="Defense", hue = 'Type 1', style = 'Generation', data=pokemon) #qualitative palette 
```

```python
sns.relplot(x="Attack", y="Defense", hue = 'Type 1', size = 'Generation', data=pokemon) #size
```

```python
#boolean condition for pd.loc
sns.relplot(x="Attack", y="Defense", hue = 'Type 1',palette="ch:r=-.55,l=.50", 
            size = 'Generation', 
            data=pokemon.loc[(pokemon['Type 1'] == 'Rock') | (pokemon['Type 1'] == 'Grass')]) #size
```

```python
sns.relplot(x="Attack", y="Defense", hue = 'Speed', style = 'Generation', data=pokemon) #sequential palette
```

```python
#information will be preserved if printed to black and white or viewed by someone who is colorblind. 
#seaborn.cubehelix_palette
sns.relplot(x="Attack", y="Defense", hue = 'Speed', palette="ch:r=-.75,l=.80", data=pokemon) #sequential palette
```

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
#the same x value could have multiple y value: when aggregation true
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

```python
#remove confidence interval
sns.relplot(x="timepoint", y="signal", ci = None,kind="line", data=fmri)
```

```python
#standard deviation instead of confidence interval
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

```python
#the same x value could have multiple y value: when aggregation false
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri)
```

```python
#colour event 
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
#line style
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
#identify subsets by the markers used at each observation, either together with the dashes or instead of them
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=True, markers=True, kind="line", data=fmri);
```

```python
#When you are working with repeated measures data (that is, you have units that were sampled multiple times), 
#you can also plot each sampling unit separately without distinguishing them through semantics. 
#This avoids cluttering the legend:
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"))
```

```python
# Working with dates
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
sns.relplot(x="Attack", y="Defense", hue = 'Type 1', col = 'Generation',col_wrap = 3, data=pokemon) #faceting using col
```

# Plotting with categorical data¶


```python
pokemon.head()
```

```python
q = sns.catplot(x = 'Type 1', y = 'Attack', data = pokemon)
q.fig.autofmt_xdate()
```

```python
#set jitter to False
w = sns.catplot(x = 'Type 1', y = 'Attack', data = pokemon, jitter = False)
w.fig.autofmt_xdate()
```

```python
#Not a useful plot. but shows how to use hue
e = sns.catplot(x="Type 1", y="Attack", hue="Generation", kind="swarm", data=pokemon)
e.fig.autofmt_xdate()
```

```python
#subseting using query method
r = sns.catplot(x="Generation", y="Attack",jitter = False, hue = 'Legendary',
            data=pokemon.query("Generation == 3 or Generation == 2"))
r.fig.autofmt_xdate()
```

```python
#Ordering the X axis values
t = sns.catplot(x = 'Legendary', y = 'Attack', order = [True,False],data = pokemon, kind = 'violin' )
```

```python
#horizontal plot
y = sns.catplot(y = 'Type 1', x = 'Attack',data = pokemon )
```

```python
#box plot
y = sns.catplot(y = 'Type 1', x = 'Attack',color = '0.7',kind = 'box',data = pokemon )
```

```python
#side by side box plot
y = sns.catplot(y = 'Type 1', x = 'Attack', hue = 'Legendary', height = 10 ,kind = 'box', data = pokemon )
```

```python
#boxen plot
sns.catplot(x = 'Legendary', y = 'Attack', order = [True,False],data = pokemon, kind = 'boxen' )
```

```python
# Combine violin and swarm plots
u = sns.catplot(x = 'Generation', y = 'Attack', kind = 'violin', data = pokemon)
sns.swarmplot(x = 'Generation', y = 'Attack', color = '0' , data = pokemon)
u.fig.autofmt_xdate()
```

```python
#bar plots
sns.catplot(x = 'Generation', y = 'Defense', hue = 'Legendary', kind = 'bar', data = pokemon)
```

```python
#Count on y axis
sns.catplot(x = 'Generation', kind = 'count',palette="ch:.25" ,data = pokemon)
```

```python
#point plots
sns.catplot(x = 'Generation',y = 'Speed', hue = 'Legendary',
            palette = {True:"g",False:"m"},markers=["^", "o"],
            linestyles =["-","--"], kind = 'point', data = pokemon)
```
